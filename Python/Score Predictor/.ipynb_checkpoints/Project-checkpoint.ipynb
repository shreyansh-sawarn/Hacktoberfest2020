```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(port, random_state = 42, test_size = 0.2)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = "median")
```

```python
imputed_train_data = pd.DataFrame(imputer.fit_transform(train_data), columns = train_data.columns)
```

```python
imputed_train_data
```

```python
X_train_data = train_data.drop(labels = "G3", axis = 1)
X_test_data = test_data.drop(labels = "G3", axis = 1)
```

```python
X_test_data
```

```python
Y_train_data = pd.DataFrame(train_data["G3"])
Y_test_data = pd.DataFrame(test_data["G3"])
```

```python
Y_test_data
```

```python
from sklearn.pipeline import Pipeline
```

```python
my_pipeline = Pipeline([
    
    ('imputer' , SimpleImputer(missing_values = np.nan, strategy = "median"))
])
```

```python
X_train = my_pipeline.fit_transform(X_train_data)
```

```python
Y_train = my_pipeline.fit_transform(Y_train_data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
#score_mse = mean_squared_error(Y_train, prediction)
#score_rmse = np.sqrt(score_mse)
#score_rmse
```

```python
from sklearn.model_selection import cross_val_score
```

```python
train_mse = cross_val_score(model, X_train, Y_train, scoring = "neg_mean_squared_error", cv = 10)
```

```python
train_rmse = np.sqrt(-train_mse)
```

```python
train_rmse
```

```python
def test(rmse):
    print("RMSE : ", rmse)
    print("MEAN : ", rmse.mean())
    print("STD DEVIATION", rmse.std())
```

```python
test(train_rmse)
```

```python
from joblib import dump, load
dump(model, 'model.joblib') 
```
